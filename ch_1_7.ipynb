```
# FROM ch_1_1.ipynb##########################
##############################################
import pandas as pd
from random import seed
from random import randrange
from csv import reader
from math import sqrt

def load_csv(filename_url):
  CSV_url=filename_url
  pdfile=pd.read_csv(CSV_url,header=None)
  dataset=pdfile.values.tolist()
  return dataset
##################

# Dataset are in https://github.com/jbrownlee/Datasets
# URL contenente dataset CSV Pima Indians dataset
#CSV_url='https://goo.gl/2tMFne'
# URL contenente dataset CSV Pima Indians dataset parziale

#CSV_url='https://raw.githubusercontent.com/stefanoridolfi/ML_From_scratch/master/pima-indians-diabetes.data_partial_40.csv'
# URL contenente dataset Iris Flowers
#CSV_url='https://goo.gl/uqoqh7'

#dataset auto insureance
CSV_url='https://raw.githubusercontent.com/stefanoridolfi/ML_From_scratch/master/auto_insurance.csv'


# Convert string column to float
def str_column_to_float(dataset, column):
  for row in dataset:
    row[column] = float(row[column].strip())

# Split a dataset into a train ad e a test set
def train_test_split(dataset,split):
  train_list=[]
  train_size=int(split*len(dataset))
  test_list=list(dataset)
  while len(train_list)  <train_size:
    index=randrange(len(test_list))
    #print("index=",index)
    elem=test_list.pop(index)
    train_list.append(elem)
  print("train",train_list,"\ntest",test_list)
  return train_list, test_list


# Calculate root mean square error
def rmse_metric(actual, predicted):
  tot_error = 0.0
  for i in range(len(actual)):
    predicted_error=((actual[i]-predicted[i])**2)
    tot_error+=predicted_error
  mean_error=tot_error/float(len(actual))
  rmse_error=sqrt(mean_error)
  return rmse_error
  

# Evaluate an algorithm using a train/test split
def evaluate_algorithm(dataset, algorithm, split, *args):
  train, test = train_test_split(dataset,split)
  print("train in evaluate",train,"\ntest in evaluate",test)
  test_set=list()
  for row in test:
    row_copy=list(row)
    row_copy[-1]=None
    test_set.append(row_copy)
  predicted=algorithm(train,test_set, *args)
  print("Predicted in evalute",predicted)
  actual=[row[-1] for row in test]
  print("actual in Evaluate",actual)
  rmse=rmse_metric(actual,predicted)
  print("Rmse in Evaluate",rmse)
  return rmse

# Calculate the mean value of a list of numbers
def mean(values):
  mean=sum(values)/float(len(values))
  return mean

#calculate variance of a list of number
def variance(values, mean):
  variance=0.0
  for x in values:
    elem=((x-mean)**2)
    variance=variance+elem 
  return variance


#calculate covariance between xand y
def covariance(x,mean_x,y,mean_y):
  covar=0.0
  for i in range(len(x)):
    elem=(x[i]-mean_x)*(y[i]-mean_y)
    covar+=elem
  return covar  

# Calculate coefficients b0 (q) and b1(m)
def coefficients(dataset):
  x=[row[0] for row in  dataset]
  y=[row[1] for row in  dataset]
  x_mean,y_mean = mean(x),mean(y)
  print("in coefficient, X_mean=",x_mean, "y_mean=",y_mean)
  b1=covariance(x,x_mean,y,y_mean)/variance(x,x_mean)
  b0=y_mean-b1*x_mean
  return[b0,b1]

# Simple linear regression algotithm
def simple_linear_regression(train,test):
  print("in simple_linear_alg: train=",train,"test=",test)
  predictions=list()
  b0, b1 = coefficients(train)
  for row in test:
    yhat=b0+b1*row[0]
    predictions.append(yhat)
  return predictions

#########################
# Test the train/test simple linear regression
seed(1)
# load and prepare data
dataset=load_csv(CSV_url)
print('Loaded data file {0} with {1} rows and {2} columns\n\n'.format(CSV_url, len(dataset),len(dataset[0])))
print("\nLista valori file CSV")
for i in range(len(dataset)):
  print("riga=",i+1, "valore",dataset[i])

split=0.6
rmse=evaluate_algorithm(dataset, simple_linear_regression, split)
print("RMSE",rmse)
print('\nRMSE: %.3f' % (rmse))
```

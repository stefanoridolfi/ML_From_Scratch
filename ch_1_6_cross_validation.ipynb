```
# FROM ch_1_1.ipynb##########################
##############################################
import pandas as pd
from random import seed
from random import randrange
from csv import reader

def load_csv(filename_url):
  CSV_url=filename_url
  pdfile=pd.read_csv(CSV_url,header=None)
  dataset=pdfile.values.tolist()
  return dataset
##################

# URL contenente dataset CSV Pima Indians dataset
#CSV_url='https://goo.gl/2tMFne'
# URL contenente dataset CSV Pima Indians dataset parziale

CSV_url='https://raw.githubusercontent.com/stefanoridolfi/ML_From_scratch/master/pima-indians-diabetes.data_partial_40.csv'
# URL contenente dataset Iris Flowers
#CSV_url='https://goo.gl/uqoqh7'


# Convert string column to float
def str_column_to_float(dataset, column):
  for row in dataset:
    row[column] = float(row[column].strip())

# Split a dataset into a train and test set
def train_test_split(dataset, split):
  train = list()
  train_size = split * len(dataset)
  dataset_copy = list(dataset)
  while len(train) < train_size:
    index = randrange(len(dataset_copy))
    elem=dataset_copy[index]
    dataset_copy.pop(index)
    train.append(elem)
  return train, dataset_copy

# Calculate accuracy percentage
def accuracy_metric(actual, predicted):
  correct = 0
  for i in range(len(actual)):
    if actual[i] == predicted[i]:
      correct += 1
  return correct / float(len(actual)) * 100.0


# Evaluate an algorithm using a train/test split
def evaluate_algorithm(dataset, algorithm, split, *args):
  train, test = train_test_split(dataset, split)
  test_set = list()
  for row in test:
    row_copy = list(row)
    row_copy[-1] = None
    test_set.append(row_copy)
  #print("test_set list dopo l'inizializzazione a None di tutte i valori",test_set)
  predicted = algorithm(train, test_set, *args)
  #predicted = algorithm(train, test, *args)
  actual = [row[-1] for row in test]
  accuracy = accuracy_metric(actual, predicted)
  return accuracy

# zero rule algorithm for classification
def zero_rule_algorithm_classification(train, test):
  output_values = [row[-1] for row in train]
  print("output_values in zero rule algorithm: lista con i valori dell'ultima colonna di tutti i record, ovvero le righe del file test",output_values)
  prediction = max(set(output_values), key=output_values.count)
  print("il valore predetto che verrÃ  usato per tutti i risultati prediction=", prediction)
  predicted = [prediction for i in range(len(test))]
  print("lista delle predizioni",predicted)
  return predicted

# Test the train/test harness
seed(1)
# load and prepare data
dataset=load_csv(CSV_url)
train, test =train_test_split(dataset, 0.6)
print('Loaded data file {0} with {1} rows and {2} columns\n\n'.format(CSV_url, len(dataset),len(dataset[0])))
#print("\ndataset[0]",dataset[0])
print("\nLista valori file CSV")
for i in range(len(dataset)):
  print("riga=",i+1, "valore",dataset[i])

print("\nLista valori file train")
for i in range(len(train)):
  print("riga=",i+1, "valore",train[i])

print("\nLista valori file test")
for i in range(len(test)):
  print("riga=",i+1, "valore",test[i])

#values_corrected = [row[-1] for row in train]
#print("valori corretti",values_corrected)
#predicted=zero_rule_algorithm_classification(train, test)
#print("lista delle predizioni",predicted)
#for i in range(len(dataset[0])):
#  str_column_to_float(dataset, i)
# evaluate algorithm
split = 0.6
accuracy = evaluate_algorithm(dataset, zero_rule_algorithm_classification, split)
print('\nAccuracy: %.3f%%' % (accuracy))
```
